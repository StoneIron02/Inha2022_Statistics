```python
import numpy as np
x = np.array([1, 3, 4, 7, 9, 12, 13, 14])
y = np.array([3.5, 2.4, 2.1, 1.3, 1.2, 2.2, 2.6, 4.2])
print(x)
print(y)
print(len(x), len(y))
```

```python
var_x = np.array([i - np.mean(x) for i in x])
var_y = np.array([i - np.mean(y) for i in y])
print(var_x)
print(var_y)
```

```python
S_xx = 0
for i in var_x:
    S_xx += i**2

S_yy = 0
for i in var_y:
    S_yy += i**2
    

S_xy = 0
for i in range(0, len(var_x)):
    S_xy += var_x[i] * var_y[i]
    
print(S_xx)
print(S_yy)
print(S_xy)
```

```python
import math

r = S_xy / (math.sqrt(S_xx) * math.sqrt(S_yy))
print(r)
```
